```python
import os

#importing libraries
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np # linear algebra
from matplotlib import pyplot as plt
import seaborn as sns

#automated feature generation tools
import featuretools as ft 
from tsfresh import extract_features
from tsfresh import extract_relevant_features
```

```python
df = pd.read_csv('./Data/1_out.csv', parse_dates=['date'])
df = df.dropna()
df["store_item"] = df["store"].astype(str) +  "," + df["item"].astype(str)
y = df["sales"]

df = df.drop(columns=["Unnamed: 0"])
df = df[["date","month","day_of_month","week_of_year","day_of_year","day_of_week","year","store_item"]]
df
```

<h1>Automated Feature Generation 
<h2>tsFresh Feature Extraction

```python
extracted_features = extract_features(df,column_id="store_item",column_sort="date")
extracted_features
```

```python
from tsfresh.utilities.dataframe_functions import impute
impute(extracted_features)
from tsfresh import select_features
filtered_features = select_features(extracted_features, y)
filtered_features
```
